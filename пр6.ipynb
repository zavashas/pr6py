Цель: необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков, а именно: линейный график, столбчатая диаграмма, круговая диаграмма, диаграмма рассеяния, гистограмма.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

titanic = pd.read_csv("titanic.csv")
```

1. Линейный график (используется наложение)
В данном примере изображен линейный график, где показаны данные 10 пассажиров: возраст, стоимость проезда, выжил или нет.
plt.legend() означает описание элементов графика (слева сверху)
plt.grid() создает сетку на нашем графике
plt.xlabel() подпись по оси X
plt.ylabel() подпись по оси Y
plt.show()как раз и делает наш график видмым, т.е отображает наш график.
На данном графике мы видим, что в основном выжили женщины, так как впервую очередь спасали их. 

```python
plt.plot(titanic["Age"].head(20), titanic["Name"].head(20), color = "blue", marker = "o", label = "Возраст")
plt.plot(titanic["Fare"].head(20), titanic["Name"].head(20),color = "m", marker = "o", label = "Стоимость проезда")
plt.plot(titanic["Survived"].head(20), titanic["Name"].head(20), color = "red", marker = "o", label = "Выжил")
plt.legend()
plt.grid()
plt.xlabel('Данные')
plt.ylabel('Пассажиры')
plt.show()
```

2. Столбчатая диаграмма
В данном примере будем использовать bar для создания столчатой диаграммы.
Данные возьмем такие: возраст пассажира, выжил он или нет.
На данной диаграмме видно, что в основном выжили пассажиры в возрасте от 15 до 35.

```python
plt.bar(titanic["Age"].head(100),titanic["Survived"].head(100), color = "rosybrown", label = "Выжил")
plt.legend()
plt.show()
```

На данной диаграмме видно, что стоимость билета 1 класса выше в 2 раза, чем 2 класс, 1 класс немного дешевле 2-го

```python
plt.bar(titanic["Pclass"].head(10),titanic["Fare"].head(10), color = "pink", label = "Цена")
plt.legend()
plt.show()
```

На данной диаграмме видно, что у пассажиров 3 класса, больше всего родителей/детей на борту

```python
plt.bar(titanic["Pclass"].head(100),titanic["Parents/Children Aboard"].head(100), color = "lightblue", label = "Родители/дети на борту")
plt.legend()
plt.show()
```

3. Круговая диаграмма.
В данном примере будем использовать pie для создания круговой диаграммы.
"a" содержит в себе информацию о том, сколько раз количество встречается в данных.
Параметр labels устанавливает подписи для каждого сектора.
Параметр autopct пописывает каждый сектр в процентных значениях
Соотношение количества братьев, сестер/супруг на борту у пассажиров
plt.title() - название диаграммы
На данной диаграмме видно, что большинство пассажиров были на борту без братьев, сестер/супруг, 23% - с.

```python
a = titanic['Siblings/Spouses Aboard'].value_counts()
plt.pie(a, labels=a.index, autopct='%1.1f%%', colors=['pink','red','m','blue','green', 'lime','purple'])
plt.title ('Соотношение пассажиров с братьями, сестрами/супргами на борту')
plt.show()
```

На данной диаграмме видно, что среди пассажиров, большинство это мужчины

```python
cl = titanic['Sex'].value_counts()
plt.pie(cl, labels=cl.index, autopct='%1.1f%%', colors=['lightblue','pink'])
plt.title ('Соотношение гендера')
plt.show()
```

4. Диаграмма рассеяния 
В данном примере будем использовать figure для создания новой фигуры с резмером 5х15 дюймов
plt.scatter() - строит точечный график, где внутри круглых скобок мы указывает данные, откуда будут браться наши данные. Там же задаем цвет точек
plt.title() - название диаграммы.

```python
plt.figure(figsize=(5,15))
plt.scatter(titanic['Pclass'].head(100),titanic['Fare'].head(100), color = 'green')
plt.title('Связь между классом и ценой за билет')
plt.xlabel('класс')
plt.ylabel('цена')
plt.show()
```

```python
plt.figure(figsize=(5,15))
plt.scatter(titanic['Sex'].head(500),titanic['Fare'].head(500), color = 'green')
plt.title('Связь между полом и ценой')
plt.xlabel('пол')
plt.ylabel('выжил/не выжил')
plt.show()
```
